# Основные определения

- *испытание* – единичное наблюдение (одно подбрасывание монетки)
- *эксперимент* — множественные испытания (подбрасывание монетки $n$ раз)
- *выборочное пространство `S`* – набор всех *возможных* элементарных исходов *испытания*, для монетки это `S = {орел, решка}` или сокращенно `S = {о, р}`. Выборочное пространство можно задать и для *эксперимента*. Так, для подбрасывания двух монет `S = {(о, о), (о, р), (р, о), (р, р)}`
- *событие `E`* – частный случай исхода *испытания* (выпадение орла при одном подбрасывании монетки `E = {о}`, выпадение нечетного числа при броске одной игральной кости» можно записать как `E = {1, 3, 5}`)
- *объединение событий* $E \cup F$ – событие происходит, если случается хотя бы одно входящее элементарное событие
- *пересечение событий* $E \cap F$  – событие происходит, если имеют место все входящие элементарные события
- *дополнение события* $\sim E$ – любое событие из *выборочного пространства* $S$, кроме события $E$
- *взаимнно исключающие события* – события, которые не могут происходить одновременно
- *независимые испытания* – исход одного события не влияет на исход другого (подбрасывание монетки)

# Перестановки и сочетания
- *перестановки P* – все возможные способы упорядочивания элементов в наборе (для набора элементов $(a, b, c)$ все перестановки следующие $(a, b, c), (a, c, b), (b, a, c), (b, c, a), (c, a, b)$ и $(c, b, a)$). То есть для перестановок важен порядок элементов. Число перестановок группы из $n$ элементов равно $n!$
- *сочетание C* – все возможные наборы элементов совокупности без учета их порядка.
- $3P2$ – число возможных перестановок двух элементов из трех, $3P2 = 6:$ $(a, b), (a, c), (b, c), (b, a), (c, a)$ и $(c, b)$.
- $3C2$ – число возможных сочетаний двух элементов из трех, $3C2 = 3:$ $(a, b), (a, c)$ и $(b, c)$

$$nPk = \dfrac{n!}{(n-k)!},$$

$$nCk = \dfrac{n!}{k!(n-k)!} = \dfrac{nPk}{k!},$$

# Вероятность
*Вероятность P события E* – величина $P(E)$ в диапазоне от 0 до 1, показывающая как часто происходит некоторое событие при повторении эксперимента. Если вероятность события равна 1, оно обязательно произойдет, если 0 – у него нет шансов случиться.

*Вероятность выборочного пространства* всегда равна 1 $P(S) = 1$. Этот же факт означает, что сумма вероятностей события и его дополнения всегда равна 1:

$$P(E) + P(\sim E) = 1.$$

Это свойство часто используется для вычислений.

Часто нужно знать вероятность события $E$ при условии, что произошло другое событие $F$. Такую вероятность $P(E|F)$ называют условной. Пример:

`P(рак легких | курильщик) > P(рак легких | некурящий).`

Две переменные независимы, если выполняется условие:

$$P(E|F) = P(E).$$
