–ü—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–µ –ø—Ä–∞–≤–∏–ª–∞ –ø–æ–º–æ–≥–∞—é—Ç –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å —Å–ø–∏—Å–æ–∫, –∫–∞–∫ –ø–æ–ª–Ω–æ–µ –¥–≤–æ–∏—á–Ω–æ–µ –¥–µ—Ä–µ–≤–æ. –¢–∞–∫–∂–µ –Ω—É–∂–Ω–æ –ø–æ–Ω–∏–º–∞—Ç—å, —á—Ç–æ —É –ª—é–±–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –µ—Å—Ç—å —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–π —ç–ª–µ–º–µ–Ω—Ç, –Ω–æ –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –¥–æ—á–µ—Ä–Ω–∏–π. –ï—Å–ª–∏ –∏–Ω–¥–µ–∫—Å `2*k` –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –∑–∞ –ø—Ä–µ–¥–µ–ª–∞–º–∏ —Å–ø–∏—Å–∫–∞, —Ç–æ —É —ç–ª–µ–º–µ–Ω—Ç–∞ –Ω–µ—Ç –¥–æ—á–µ—Ä–Ω–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤. –ï—Å–ª–∏ `2*k` —è–≤–ª—è–µ—Ç—Å—è –¥–æ–ø—É—Å—Ç–∏–º—ã–º –∏–Ω–¥–µ–∫—Å–æ–º, –∞ `2*k+1` ‚Äì –Ω–µ—Ç, —Ç–æ —É —ç–ª–µ–º–µ–Ω—Ç–∞ –µ—Å—Ç—å –ª–∏—à—å –æ–¥–∏–Ω –ø–æ—Ç–æ–º–æ–∫.

–°–≤–æ–π—Å—Ç–≤–æ –∫—É—á–∏ –ø–æ–¥—Ä–∞–∑—É–º–µ–≤–∞–µ—Ç, —á—Ç–æ –µ—Å–ª–∏ `h` —è–≤–ª—è–µ—Ç—Å—è –∫—É—á–µ–π, —Ç–æ —Å–ª–µ–¥—É—é—â–µ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ –±—É–¥–µ—Ç –ª–æ–∂–Ω—ã–º:

```python
h[k] <= h[2*k + 1] and h[k] <= h[2*k + 2]
```

–í—ã—Ä–∞–∂–µ–Ω–∏–µ –º–æ–∂–µ—Ç –≤—ã–∑–≤–∞—Ç—å –∏—Å–∫–ª—é—á–µ–Ω–∏–µ `IndexError`, –µ—Å–ª–∏ –∏–Ω–¥–µ–∫—Å –≤—ã—à–µ–ª –∑–∞ –ø—Ä–µ–¥–µ–ª—ã —Å–ø–∏—Å–∫–∞, –Ω–æ –≤—ã—á–∏—Å–ª—è–µ–º–æ–µ –ª–æ–≥–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∏–∫–æ–≥–¥–∞ –Ω–µ –±—É–¥–µ—Ç `False`.

–¢–æ –µ—Å—Ç—å –≤ —Ç–µ—Ä–º–∏–Ω–∞—Ö —Å–ø–∏—Å–∫–æ–≤: k-–π —ç–ª–µ–º–µ–Ω—Ç –≤—Å–µ–≥–¥–∞ –º–µ–Ω—å—à–µ, —á–µ–º –∫–∞–∂–¥—ã–π –∏–∑ –ø–∞—Ä—ã —ç–ª–µ–º–µ–Ω—Ç–æ–≤, —Å—Ç–æ—è—â–∏—Ö –≤—Å–ª–µ–¥ –∑–∞ 2k-–º —ç–ª–µ–º–µ–Ω—Ç–æ–º.

–ü—Ä–∏–º–µ—Ä —Å–ø–∏—Å–∫–∞, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–µ–≥–æ —Å–≤–æ–π—Å—Ç–≤—É –∫—É—á–∏:

![](https://files.realpython.com/media/heap-list.2db422e53ead.png)

–°—Ç—Ä–µ–ª–∫–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω—ã –æ—Ç —ç–ª–µ–º–µ–Ω—Ç–∞ —Å –∏–Ω–¥–µ–∫—Å–æ–º `k` –∫ —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å –∏–Ω–¥–µ–∫—Å–∞–º–∏ `2 * k + 1` –∏ `2 * k + 2`. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ —Å–ø–∏—Å–∫–µ Python –∏–º–µ–µ—Ç –∏–Ω–¥–µ–∫—Å 0, –ø–æ—ç—Ç–æ–º—É –¥–≤–µ —Å—Ç—Ä–µ–ª–∫–∏ —É–∫–∞–∑—ã–≤–∞—é—Ç –Ω–∞ –∏–Ω–¥–µ–∫—Å—ã 1 –∏ 2.

# –ë–∞–∑–æ–≤—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏

–í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –º–Ω–æ–≥–∏—Ö –¥—Ä—É–≥–∏—Ö –º–æ–¥—É–ª–µ–π Python,  `heapq` –Ω–µ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∫–ª–∞—Å—Å, –∞ —Ä–∞–±–æ—Ç–∞–µ—Ç —Å–æ —Å–ø–∏—Å–∫–∞–º–∏ –Ω–∞–ø—Ä—è–º—É—é. 

–û–±—ã—á–Ω–æ, –∫–∞–∫ –≤ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –ø–æ—á—Ç—ã, —ç–ª–µ–º–µ–Ω—Ç—ã –≤—Å—Ç–∞–≤–ª—è—é—Ç—Å—è –≤ –∫—É—á—É –æ–¥–∏–Ω –∑–∞ –¥—Ä—É–≥–∏–º, –Ω–∞—á–∏–Ω–∞—è —Å –ø—É—Å—Ç–æ–π –∫—É—á–∏. –û–¥–Ω–∞–∫–æ, –µ—Å–ª–∏ —É–∂–µ –µ—Å—Ç—å —Å–ø–∏—Å–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∏—Ö –º–æ–∂–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ `heapify()`:

````python
import heapq
a = [3, 5, 1, 2, 6, 8, 7]
heapq.heapify(a)
a
````

–•–æ—Ç—è 7 –∏–¥–µ—Ç –ø–æ—Å–ª–µ 8, –ª–µ–≥–∫–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —á—Ç–æ —Å–ø–∏—Å–æ–∫ –ø–æ–¥—á–∏–Ω—è–µ—Ç—Å—è —Å–≤–æ–π—Å—Ç–≤—É –∫—É—á–∏. –ù–∞–ø—Ä–∏–º–µ—Ä, –∑–Ω–∞—á–µ–Ω–∏–µ `a[2]`, —Ä–∞–≤–Ω–æ–µ 3, –º–µ–Ω—å—à–µ –∑–Ω–∞—á–µ–Ω–∏—è `a[2*2 + 2]`, —Ä–∞–≤–Ω–æ–≥–æ 7.

`heapify()` –∏–∑–º–µ–Ω—è–µ—Ç —Å–ø–∏—Å–æ–∫ –Ω–∞ –º–µ—Å—Ç–µ, –Ω–æ –Ω–µ —Å–æ—Ä—Ç–∏—Ä—É–µ—Ç –µ–≥–æ. –ß—Ç–æ–±—ã —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—Ç—å –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–º—É —Å–≤–æ–π—Å—Ç–≤—É –∫—É—á–∞ –Ω–µ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∞. –í —Ç–æ –∂–µ –≤—Ä–µ–º—è –ª—é–±–æ–π –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è–µ—Ç —Å–≤–æ–π—Å—Ç–≤—É –∫—É—á–∏ –∏ –∑–∞–ø—É—Å–∫ `heapify()` –Ω–∞ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–º —Å–ø–∏—Å–∫–µ –Ω–µ –∏–∑–º–µ–Ω–∏—Ç –ø–æ—Ä—è–¥–æ–∫ —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

–î—Ä—É–≥–∏–µ –æ—Å–Ω–æ–≤–Ω—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ –≤ –º–æ–¥—É–ª–µ Python `heapq` –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—é—Ç, —á—Ç–æ —Å–ø–∏—Å–æ–∫ —É–∂–µ —è–≤–ª—è–µ—Ç—Å—è –∫—É—á–µ–π. –ü–æ–ª–µ–∑–Ω–æ –æ—Ç–º–µ—Ç–∏—Ç—å, —á—Ç–æ –∫—É—á–µ–π –≤—Å–µ–≥–¥–∞ —è–≤–ª—è—é—Ç—Å—è –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –∏–ª–∏ —Å–ø–∏—Å–æ–∫ –µ–¥–∏–Ω–∏—á–Ω–æ–π –¥–ª–∏–Ω—ã.

–ü–æ—Å–∫–æ–ª—å–∫—É –∫–æ—Ä–µ–Ω—å –¥–µ—Ä–µ–≤–∞ —è–≤–ª—è–µ—Ç—Å—è –ø–µ—Ä–≤—ã–º —ç–ª–µ–º–µ–Ω—Ç–æ–º, –Ω–∞–º –Ω–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —á—Ç–µ–Ω–∏—è –Ω–∞–∏–º–µ–Ω—å—à–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ ‚Äì –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –∏–∑–≤–ª–µ—á—å –µ–≥–æ –ø–æ –∏–Ω–¥–µ–∫—Å—É `a[0]`. –ß—Ç–æ–±—ã –≤—ã—Ç–æ–ª–∫–Ω—É—Ç—å –Ω–∞–∏–º–µ–Ω—å—à–∏–π —ç–ª–µ–º–µ–Ω—Ç –ø—Ä–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–∏ —Å–≤–æ–π—Å—Ç–≤–∞ –∫—É—á–∏, –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è `heappop()`:

````python
root = heapq.heappop(a)
print(root, a)
````

–§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç, 1. –°–∞–º–∞ –∫—É—á–∞ –∏–∑–º–µ–Ω—è–µ—Ç—Å—è, –Ω–æ —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è —Å–≤–æ–π—Å—Ç–≤–æ –∫—É—á–∏: –Ω–∞–ø—Ä–∏–º–µ—Ä, `a[1]` —Ä–∞–≤–Ω–æ 5, `–∞[1*2 + 2]` —Ä–∞–≤–Ω–æ 6.

–ú–æ–¥—É–ª—å Python `heapq` —Ç–∞–∫–∂–µ –≤–∫–ª—é—á–∞–µ—Ç –≤ —Å–µ–±—è `heappush()` –¥–ª—è –ø–æ–º–µ—â–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –∫—É—á—É –ø—Ä–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–∏ —Å–≤–æ–π—Å—Ç–≤–∞ –∫—É—á–∏.

````python
heapq.heappush(a, 4)
print(a)
````

````python
heapq.heappop(a)
````

````python
heapq.heappop(a)
````

````python
heapq.heappop(a)
````

–ú—ã –¥–æ–±–∞–≤–∏–ª–∏ –≤ –∫—É—á—É 4, –Ω–æ –∑–∞—Ç–µ–º –≤—ã—Ç–æ–ª–∫–Ω—É–ª–∏ –∏–∑ –Ω–µ–µ —Ç—Ä–∏ —ç–ª–µ–º–µ–Ω—Ç–∞. –ö–∞–∫ –º–æ–∂–Ω–æ –≤–∏–¥–µ—Ç—å, –≤—Å–µ–≥–¥–∞ –≤—ã—Ç–∞–ª–∫–∏–≤–∞–µ—Ç—Å—è –Ω–∞–∏–º–µ–Ω—å—à–∏–π —ç–ª–µ–º–µ–Ω—Ç.

–ú–æ–¥—É–ª—å Python `heapq` –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –µ—â–µ –¥–≤–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:
* `heapreplace()` —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ `heappop()` —Å –ø–æ—Å–ª–µ–¥—É—é—â–∏–º `heappush()`. 
* `heappushpop()` —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ `heappush()` c –ø–æ—Å–ª–µ–¥—É—é—â–∏–º `heappop()`.

–≠—Ç–∞ –ø–∞—Ä–∞ –º–µ—Ç–æ–¥–æ–≤ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–µ–µ, —á–µ–º –≤—Ö–æ–¥—è—â–∏–µ –≤ –Ω–∏—Ö —Ñ—É–Ω–∫—Ü–∏–∏, –∑–∞–ø—É—â–µ–Ω–Ω—ã–µ –ø–æ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏.

# –í—ã—Å–æ–∫–æ—É—Ä–æ–≤–Ω–µ–≤—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏

–ü–æ—Å–∫–æ–ª—å–∫—É –æ—á–µ—Ä–µ–¥–∏ —Å –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–º —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π, –º–æ–¥—É–ª—å `heapq` –∏–º–µ–µ—Ç –≥–æ—Ç–æ–≤—É—é —Ñ—É–Ω–∫—Ü–∏—é `merge()` –¥–ª—è –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –æ—á–µ—Ä–µ–¥–µ–π. –§—É–Ω–∫—Ü–∏—è `merge()` –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç, —á—Ç–æ –≤—Ö–æ–¥–Ω—ã–µ –∏—Ç–µ—Ä–∏—Ä—É–µ–º—ã–µ –æ–±—ä–µ–∫—Ç—ã —É–∂–µ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω—ã, –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∏—Ç–µ—Ä–∞—Ç–æ—Ä, –∞ –Ω–µ —Å–ø–∏—Å–æ–∫.

–í –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–∏–º–µ—Ä–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è `merge()` —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é –ø–ª–∞–Ω–∏—Ä–æ–≤—â–∏–∫–∞ —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω–æ–π –ø–æ—á—Ç—ã:

````python
import datetime

def email(frequency, details):
    current = datetime.datetime.now()
    while True:
        current += frequency
        yield current, details

fast_email = email(datetime.timedelta(minutes=15), "fast email")
slow_email = email(datetime.timedelta(minutes=40), "slow email")

unified = heapq.merge(fast_email, slow_email)
````

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è `merge()` –≤ —ç—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ —è–≤–ª—è—é—Ç—Å—è –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–º–∏ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞–º–∏. –í–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –ø—Ä–∏—Å–≤–æ–µ–Ω–Ω–æ–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π `unified` —è–≤–ª—è–µ—Ç—Å—è –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–º –∏—Ç–µ—Ä–∞—Ç–æ—Ä–æ–º. –≠—Ç–æ—Ç –∏—Ç–µ—Ä–∞—Ç–æ—Ä –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω—ã–µ –ø–∏—Å—å–º–∞, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω—ã –≤ –ø–æ—Ä—è–¥–∫–µ –±—É–¥—É—â–∏—Ö –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –º–µ—Ç–æ–∫.

–î–ª—è –æ—Ç–ª–∞–¥–∫–∏ –∏ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏ —Å–ª–∏—è–Ω–∏—è –∫–æ–¥–∞ –≤—ã–≤–µ–¥–µ–º –Ω–∞ —ç–∫—Ä–∞–Ω –¥–µ—Å—è—Ç—å –ø–µ—Ä–≤—ã—Ö –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –ø–∏—Å–µ–º:

````python
for _ in range(10):
    print(next(unified))
````

–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ `fast email` –∏–¥—É—Ç –∫–∞–∂–¥—ã–µ 15 –º–∏–Ω—É—Ç, –∞ `low email` –∫–∞–∂–¥—ã–µ 40 –º–∏–Ω—É—Ç, –ø—Ä–∏ —ç—Ç–æ–º —ç–ª–µ–∫—Ç—Ä–æ–Ω–Ω—ã–µ –ø–∏—Å—å–º–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ —á–µ—Ä–µ–¥—É—é—Ç—Å—è –∏ –∏–¥—É—Ç –≤ –ø–æ—Ä—è–¥–∫–µ –∏—Ö –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –º–µ—Ç–æ–∫. –ü—Ä–∏ —ç—Ç–æ–º —Ñ—É–Ω–∫—Ü–∏—è `merge()` –Ω–µ —á–∏—Ç–∞–µ—Ç –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Ü–µ–ª–∏–∫–æ–º, –∞ —Ä–∞–±–æ—Ç–∞–µ—Ç –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–∏.

–î—Ä—É–≥–æ–π –ø—Ä–∏–º–µ—Ä ‚Äì –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π —Å—Ç—Ä–æ–∫ –∏–∑ —Ä–∞–∑–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤ –∂—É—Ä–Ω–∞–ª–∏—Ä–æ–≤–∞–Ω–∏—è. –î–∞–∂–µ –µ—Å–ª–∏ —Ñ–∞–π–ª—ã –±–æ–ª—å—à–∏–µ, –æ–ø–µ—Ä–∞—Ü–∏—è –∑–∞–Ω–∏–º–∞–µ—Ç —Ä–∞–∑—É–º–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ä–µ–º–µ–Ω–∏. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ç–∞–∫–∂–µ –¥—Ä—É–≥–∏–µ –∑–∞–¥–∞—á–∏.

–ö–∞—Ä—Ç–∞ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Å—Ç—Ä–æ–∫—É –≤ —Ç—Ä–æ–π–Ω—ã—Ö –∫–∞–≤—ã—á–∫–∞—Ö, –æ—Ç–æ–±—Ä–∞–∂–∞—é—â—É—é –æ–±–ª–∞—Å—Ç—å, –≤ –∫–æ—Ç–æ—Ä–æ–π —Ä–æ–±–æ—Ç –º–æ–∂–µ—Ç –¥–≤–∏–≥–∞—Ç—å—Å—è, –∞ —Ç–∞–∫–∂–µ –ø—Ä–µ–ø—è—Ç—Å—Ç–≤–∏—è (–≤ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ —Å –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–π —Å—Ç–∞—Ç—å–µ–π –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –º—ã –∑–∞–º–µ–Ω–∏–ª–∏ —Ç–æ—á–∫–∏ –∏ –∫—Ä–µ—Å—Ç–∏–∫–∏ –Ω–∞ —ç–º–æ–¥–∑–∏ ‚Äî –ø—Ä–∏–º. –ø–µ—Ä.).

–•–æ—Ç—è –≤ –±–æ–ª–µ–µ —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ–º —Å—Ü–µ–Ω–∞—Ä–∏–∏ –∫–∞—Ä—Ç–∞ —Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∏–∑ —Ñ–∞–π–ª–∞, –¥–ª—è —Ü–µ–ª–µ–π –æ–±—É—á–µ–Ω–∏—è –ø—Ä–æ—â–µ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –Ω–∞–≥–ª—è–¥–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –≤ –∫–æ–¥–µ. –ö–æ–¥ –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–∞ –ª—é–±–æ–π –∫–∞—Ä—Ç–µ, –Ω–æ –¥–ª—è –æ—Ç–ª–∞–¥–∫–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è –ø—Ä–æ—Å—Ç—ã–º –ø—Ä–∏–º–µ—Ä–æ–º.

–û–ø—Ä–µ–¥–µ–ª–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –∫–∞—Ä—Ç—É —Ç–∞–∫, —á—Ç–æ–±—ã –µ–µ –±—ã–ª–æ –ø—Ä–æ—â–µ –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º–Ω–æ:

````python
def parse_map(m):
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–æ—Å—Ç—Ä–æ—á–Ω—É—é –∫–∞—Ä—Ç—É, –ø–æ–∑–∏—Ü–∏–∏ –Ω–∞—á–∞–ª–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç
    –∏ –ø–æ–∑–∏—Ü–∏—é –º–µ—Å—Ç–∞."""
    lines = m.splitlines()
    
     # –Ω–∞—á–∞–ª–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç (—Ç–æ—á–∫–∞ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–∏—è)
    origin = 0, 0 
    
     # –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –º–µ—Å—Ç–∞ –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è
    destination = len(lines[-1]) - 1, len(lines) - 1 
    return lines, origin, destination
````

–°–ø–∏—Å–æ–∫ `lines` –º–æ–∂–Ω–æ –∏–Ω–¥–µ–∫—Å–∏—Ä–æ–≤–∞—Ç—å —Å –ø–æ–º–æ—â—å—é –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç (x, y) ‚Äì –≤—ã—Ä–∞–∂–µ–Ω–∏–µ `lines[y][x]` –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ–∑–∏—Ü–∏–∏ –≤ –≤–∏–¥–µ –æ–¥–Ω–æ–≥–æ –∏–∑ –¥–≤—É—Ö —Å–∏–º–≤–æ–ª–æ–≤:

* ‚ö™ ‚Äì –ø—É—Å—Ç–∞—è –ø–æ–∑–∏—Ü–∏—è
* üå≤ ‚Äì –ø–æ–∑–∏—Ü–∏—è –ø—Ä–µ–ø—è—Ç—Å—Ç–≤–∏—è

–°–ª–µ–¥—É—é—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –º–æ–∂–µ—Ç –ª–∏ —Ä–æ–±–æ—Ç –∑–∞–Ω—è—Ç—å –ø–æ–∑–∏—Ü–∏—é (x, y) —Å —É—á–µ—Ç–æ–º —Å–æ–¥–µ—Ä–∂–∞–Ω–∏—è –∫–∞—Ä—Ç—ã –∏ –µ–µ –≥—Ä–∞–Ω–∏—Ü:

````python
def is_valid(lines, position):
    """–ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –Ω–µ –≤—ã—Ö–æ–¥–∏—Ç –ª–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ position –∑–∞ –≥—Ä–∞–Ω–∏—Ü—ã –∫–∞—Ä—Ç—ã
    –∏ –¥–æ—Å—Ç—É–ø–Ω–∞ –ª–∏  position –¥–ª—è –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è"""
    x, y = position
    if not (0 <= y < len(lines) and 0 <= x < len(lines[y])):
        return False
    if lines[y][x] != "‚ö™":
        return False
    return True
````

–û–ø–∏—Å–∞–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é `is_valid()` –ø—Ä–∏–º–µ–Ω–∏–º –≤ —Å–ª–µ–¥—É—é—â–µ–π —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Å–æ—Å–µ–¥–µ–π —Ç–µ–∫—É—â–µ–π –ø–æ–∑–∏—Ü–∏–∏:

````python
def get_neighbors(lines, current):
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–æ—Ä–¥–∏–∞–Ω—Ç—ã —Å–æ—Å–µ–¥–µ–π —Ç–µ–∫—É—â–µ–π –ø–æ–∑–∏—Ü–∏–∏"""
    x, y = current
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            position = x + dx, y + dy
            if is_valid(lines, position):
                yield position
````

–ü–æ—Å–ª–µ–¥–Ω—è—è –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –±—É–¥–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –ø—É—Ç–∏:

````python
def get_shorter_paths(tentative, positions, through):
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–æ–∑–∏—Ü–∏–∏, –ø—Ä–∏ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏–∏ —á–µ—Ä–µ–∑ –∫–æ—Ç–æ—Ä—ã–µ
    –ø—É—Ç—å –∫–æ—Ä–æ—á–µ —Ç–µ–∫—É—â–µ–≥–æ –ø—É—Ç–∏"""
    path = tentative[through] + [through]
    for position in positions:
        if position in tentative and len(tentative[position]) <= len(path):
            continue
        yield position, path
````

* `positions` ‚Äî –∏—Ç–µ—Ä–∏—Ä—É–µ–º—ã–π –æ–±—ä–µ–∫—Ç –ø–æ–∑–∏—Ü–∏–π, –¥–æ –∫–æ—Ç–æ—Ä—ã—Ö –º—ã —Ö–æ—Ç–∏–º —Å–æ–∫—Ä–∞—Ç–∏—Ç—å –ø—É—Ç—å

–ü–æ–∑–∏—Ü–∏—è –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ `certain`, –µ—Å–ª–∏ –º—ã —É–≤–µ—Ä–µ–Ω—ã, —á—Ç–æ —Å–∞–º—ã–π –∫–æ—Ä–æ—Ç–∫–∏–π –∏–∑ –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –ø—É—Ç–µ–π ‚Äì —Å–∞–º—ã–π –∫–æ—Ä–æ—Ç–∫–∏–π –∏–∑ –≤–æ–∑–º–æ–∂–Ω—ã—Ö. –ï—Å–ª–∏ –≤ `certain` –æ–∫–∞–∑–∞–ª—Å—è –ø—É–Ω–∫—Ç –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è, –∑–Ω–∞—á–∏—Ç, –Ω–∞–π–¥–µ–Ω –∏—Å–∫–æ–º—ã–π –ø—É—Ç—å.

–ö—É—á–∞ `candidates` —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–∞ –ø–æ –¥–ª–∏–Ω–µ —Å–∞–º–æ–≥–æ –∫–æ—Ä–æ—Ç–∫–æ–≥–æ –∏–∑–≤–µ—Å—Ç–Ω–æ–≥–æ –ø—É—Ç–∏ –∏ —É–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–π –º–æ–¥—É–ª—è heapq.

–ù–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –º—ã —Å–º–æ—Ç—Ä–∏–º –Ω–∞ –ø–æ–∑–∏—Ü–∏—é-–∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –∏ –∫—Ä–∞—Ç—á–∞–π—à–∏–π –∏–∑–≤–µ—Å—Ç–Ω—ã–π –ø—É—Ç—å. –ò–∑ –∫—É—á–∏ `candidates` –≤—ã—Ç–∞–ª–∫–∏–≤–∞–µ—Ç—Å—è —ç–ª–µ–º–µ–Ω—Ç —Å –ø–æ–º–æ—â—å—é `heappop()`. –î–æ —ç—Ç–æ–π –ø–æ–∑–∏—Ü–∏–∏-–∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –±–æ–ª–µ–µ –∫–æ—Ä–æ—Ç–∫–æ–≥–æ –ø—É—Ç–∏ ‚Äì –≤—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø—É—Ç–∏ –ø—Ä–æ—Ö–æ–¥—è—Ç —á–µ—Ä–µ–∑ –¥—Ä—É–≥–∏–µ —É–∑–ª—ã, –∏ –≤—Å–µ –æ–Ω–∏ –¥–ª–∏–Ω–Ω–µ–µ. –ò–∑-–∑–∞ —ç—Ç–æ–≥–æ —Ç–µ–∫—É—â–∏–π –∫–∞–Ω–¥–∏–¥–∞—Ç –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç–º–µ—á–µ–Ω –∫–∞–∫ `certain`.

–ó–∞—Ç–µ–º –º—ã –ø—Ä–æ—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º –≤—Å–µ—Ö –Ω–µ–ø–æ—Å–µ—â—ë–Ω–Ω—ã—Ö —Å–æ—Å–µ–¥–µ–π –∏, –µ—Å–ª–∏ –ø—Ä–æ—Ö–æ–¥ —á–µ—Ä–µ–∑ —Ç–µ–∫—É—â—É—é –ø–æ–∑–∏—Ü–∏—é —è–≤–ª—è–µ—Ç—Å—è —É–ª—É—á—à–µ–Ω–∏–µ–º, –º—ã –¥–æ–±–∞–≤–ª—è–µ–º –∏—Ö –≤ –∫—É—á—É `candidates`, –∏—Å–ø–æ–ª—å–∑—É—è `heappush()`.

–û–ø–∏—Å–∞–Ω–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–µ–∞–ª–∏–∑—É–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è `find_path()`:

````python
def find_path(m):
    lines, origin, destination = parse_map(m)
    tentative = {origin: []}
    candidates = [(0, origin)]
    certain = set()
    while destination not in certain and len(candidates) > 0:
        _ignored, current = heapq.heappop(candidates)
        if current in certain:
            continue
        certain.add(current)
        neighbors = set(get_neighbors(lines, current)) - certain
        shorter = get_shorter_paths(tentative, neighbors, current)
        for neighbor, path in shorter:
            tentative[neighbor] = path
            heapq.heappush(candidates, (len(path), neighbor))
    if destination in tentative:
        return tentative[destination] + [destination]
    else:
        raise ValueError("no path")
````

–§—É–Ω–∫—Ü–∏—è `find_path()` –ø–æ–ª—É—á–∞–µ—Ç –∫–∞—Ä—Ç—É –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø—É—Ç—å –æ—Ç –Ω–∞—á–∞–ª–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –∫ –ø—É–Ω–∫—Ç—É –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è –≤ –≤–∏–¥–µ —Å–ø–∏—Å–∫–∞ –ø–æ–∑–∏—Ü–∏–π. –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –Ω–µ–º–Ω–æ–≥–æ –¥–ª–∏–Ω–Ω–µ–µ –∏ —Å–ª–æ–∂–Ω–µ–µ —Ç–µ—Ö, —á—Ç–æ –º—ã –ø–∏—Å–∞–ª–∏ —Ä–∞–Ω—å—à–µ ‚Äì –¥–∞–≤–∞–π—Ç–µ –ø—Ä–æ–±–µ–∂–∏–º—Å—è –ø–æ—Å—Ç—Ä–æ—á–Ω–æ:

* –í –ø–µ—Ä–≤—ã—Ö —Ç—Ä–µ—Ö —Å—Ç—Ä–æ—á–∫–∞—Ö –ø–æ—Å–ª–µ –æ–±—ä—è–≤–ª–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –º—ã –æ–±—ä—è–≤–ª—è–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥–µ—Ç –ø—Ä–æ—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –∏ –æ–±–Ω–æ–≤–ª—è—Ç—å —Ü–∏–∫–ª `while`. –ò—Å—Ö–æ–¥–Ω—ã–π –ø—É—Ç—å –∏–º–µ–µ—Ç –¥–ª–∏–Ω—É 0 ‚Äì —ç—Ç–æ –ø—É—Ç—å –æ—Ç –Ω–∞—á–∞–ª–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –≤ –Ω–µ–≥–æ –∂–µ. 
* –î–∞–ª–µ–µ –∏–¥–π—Ç —Ü–∏–∫–ª `while`. –ï—Å–ª–∏ –ø–æ–∑–∏—Ü–∏–π-–∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –Ω–µ—Ç, —Ç–æ –∏ –ø—É—Ç—å —Å–æ–∫—Ä–∞—Ç–∏—Ç—å –Ω–µ–ª—å–∑—è. –ï—Å–ª–∏ –ø—É–Ω–∫—Ç –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ `certain`, –ø—É—Ç—å –Ω–∞–π–¥–µ–Ω.
* –ü–µ—Ä–≤—ã–µ —á–µ—Ç—ã—Ä–µ —Å—Ç—Ä–æ–∫–∏ –ø–æ—Å–ª–µ –æ–±—ä—è–≤–ª–µ–Ω–∏—è —Ü–∏–∫–ª–∞ –ø–æ–ª—É—á–∞—é—Ç –ø–æ–∑–∏—Ü–∏—é-–∫–∞–Ω–¥–∏–¥–∞—Ç–∞ —Å –ø–æ–º–æ—â—å—é `heappop()`, –ø—Ä–æ–ø—É—Å–∫–∞—é—Ç –∏—Ç–µ—Ä–∞—Ü–∏—é, –µ—Å–ª–∏ –ø–æ–∑–∏—Ü–∏—è-–∫–∞–Ω–¥–∏–¥–∞—Ç —É–∂–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ `certain` –∏–ª–∏, –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ, –¥–æ–±–∞–≤–ª—è—é—Ç –µ—ë –≤ `certain`. –≠—Ç–æ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç, —á—Ç–æ –∫–∞–∂–¥–∞—è –ø–æ–∑–∏—Ü–∏—è-–∫–∞–Ω–¥–∏–¥–∞—Ç –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Å—è —Ü–∏–∫–ª–æ–º –Ω–µ –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ —Ä–∞–∑–∞.
* –°–ª–µ–¥—É—é—â–∏–µ —Å—Ç—Ä–æ–∫–∏ —Ü–∏–∫–ª–∞ –∏—â—É—Ç –±–æ–ª–µ–µ –∫–æ—Ä–æ—Ç–∫–∏–µ –ø—É—Ç–∏ –∫ —Å–æ—Å–µ–¥–Ω–∏–º –ø–æ–∑–∏—Ü–∏—è–º –∏ –æ–±–Ω–æ–≤–ª—è—é—Ç —Å–ª–æ–≤–∞—Ä—å  `tentative` –∏ –∫—É—á—É `candidates`, –∏—Å–ø–æ–ª—å–∑—É—è `get_neighbors()` –∏ `get_shorter_paths()`.
* –£—Å–ª–æ–≤–Ω–∞—è –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏—è –ø–æ—Å–ª–µ —Ü–∏–∫–ª–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç. –•–æ—Ç—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø—É—Ç–µ–π –±–µ–∑ –∫–æ–Ω–µ—á–Ω–æ–π –ø–æ–∑–∏—Ü–∏–∏ —Å–¥–µ–ª–∞–ª–æ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –±–æ–ª–µ–µ –ø—Ä–æ—Å—Ç–æ–π, –≤ —Å–∞–º–æ–º –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–µ –ª–æ–≥–∏—á–Ω–æ –≤–µ—Ä–Ω—É—Ç—å –ø—É—Ç—å —Ü–µ–ª–∏–∫–æ–º –≤–º–µ—Å—Ç–µ —Å –ø—É–Ω–∫—Ç–æ–º –Ω–∞–∑–Ω–∞—á–µ–Ω–∏—è. –ï—Å–ª–∏ –ø—É—Ç—å –Ω–µ –Ω–∞–π–¥–µ–Ω, –≤—ã–∑—ã–≤–∞–µ–º –∏—Å–∫–ª—é—á–µ–Ω–∏–µ.

# –ö–æ–¥ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞

–ò—Ç–∞–∫, —Ä–æ–±–æ—Ç—É —É–∂–µ –≤—Å—ë –ø–æ–Ω—è—Ç–Ω–æ. –ù–æ —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞–≥–ª—è–¥–Ω–µ–µ –¥–ª—è –ª—é–¥–µ–π, —Ö–æ—Ä–æ—à–æ –±—ã –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º. –ù–∞–ø–∏—à–µ–º –¥–ª—è —ç—Ç–æ–≥–æ —Ñ—É–Ω–∫—Ü–∏—é, –æ—Ç–æ–±—Ä–∞–∂–∞—é—â—É—é –ø—É—Ç—å —Ä–æ–±–æ—Ç–∞.

````python
def show_path(path, m):
    lines = m.splitlines()
    for x, y in path:
        lines[y] = lines[y][:x] + "ü§ñ" + lines[y][x+1:]
    return "\n".join(lines) + "\n"

path = find_path(m)
print(show_path(path, m))
````

–ü–æ–¥–æ–±–Ω—ã–µ –ø—Ä–æ–±–ª–µ–º—ã —Å –ø–æ–∏—Å–∫–æ–º –ø—É—Ç–∏, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ —Ä–µ—à–∏—Ç—å —Å –ø–æ–º–æ—â—å—é –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–≥–æ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è –∏ –æ—á–µ—Ä–µ–¥–µ–π —Å –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–∞–º–∏, –Ω–µ—Ä–µ–¥–∫–æ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –Ω–∞ —Å–æ–±–µ—Å–µ–¥–æ–≤–∞–Ω–∏—è—Ö –∏ —Å–æ—Ä–µ–≤–Ω–æ–≤–∞–Ω–∏—è—Ö –ø–æ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—é. –ù–∞–ø—Ä–∏–º–µ—Ä, –Ω–∞ Advent of Code 2019 –≥–æ–¥–∞  [—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â—É—é –∑–∞–¥–∞—á—É](https://adventofcode.com/2019/day/18) –±—ã–ª–æ –º–æ–∂–Ω–æ —Ä–µ—à–∏—Ç—å —Å –ø–æ–º–æ—â—å—é –æ–ø–∏—Å–∞–Ω–Ω—ã—Ö –∑–¥–µ—Å—å –º–µ—Ç–æ–¥–æ–≤.

–¢–µ–ø–µ—Ä—å –≤—ã –∑–Ω–∞–µ—Ç–µ –æ —Ç–æ–º, –∫–∞–∫ —Ä–µ–∞–ª–∏–∑—É—é—Ç—Å—è —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö –∫—É—á–∏ –≤ –º–æ–¥—É–ª–µ Python heapq –∏ –¥–ª—è –∫–∞–∫–∏—Ö –∑–∞–¥–∞—á –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è –æ–Ω–∏ –ø–æ–ª–µ–∑–Ω—ã. –ì–ª–∞–≤–Ω–æ–µ –ø–æ–Ω–∏–º–∞—Ç—å, —á—Ç–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å —ç—Ç–æ–≥–æ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–∞ –ø—Ä–æ—è–≤–ª—è–µ—Ç—Å—è –Ω–µ –≤ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ–π —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–µ –∑–Ω–∞—á–µ–Ω–∏–π, –∞ –≤ –ø–æ–∏—Å–∫–µ —ç–∫—Å—Ç—Ä–µ–º–∞–ª—å–Ω—ã—Ö –∏ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∑–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è –∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π –æ–±—Ä–∞–±–æ—Ç–∫–µ –æ—á–µ—Ä–µ–¥–µ–π —Å –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–æ–º. –§—É–Ω–∫—Ü–∏–π –≤ –º–æ–¥—É–ª–µ –Ω–µ–º–Ω–æ–≥–æ, –∞ –≤—Å—è –Ω–µ–æ–±—Ö–æ–¥–∏–º–∞—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –∏ –ø—Ä–∏–º–µ—Ä—ã –¥–æ—Å—Ç—É–ø–Ω—ã –≤ [–¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏](https://docs.python.org/3/library/heapq.html).
